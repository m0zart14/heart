```python
import pandas as pd
import numpy as np
from sklearn.cluster import DBSCAN, AgglomerativeClustering, KMeans
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
%matplotlib inline
```

```python
df = pd.read_csv("heart.csv")
```

```python
df.head()
```

```python
dbscan = DBSCAN(eps=6, min_samples=2)
```

```python
dbscan_cls = dbscan.fit_predict(df.values)
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)
```

```python
kmeans_cls = kmeans.fit_predict(df.values)
```

```python
ak = AgglomerativeClustering(n_clusters=5)
```

```python
ak_cls = ak.fit_predict(df.values)
```

```python
pca = PCA(n_components=2, random_state=42)
```

```python
coords = pca.fit_transform(df)
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(coords[:, 0], coords[:, 1], c=dbscan_cls)
plt.title("CLS DBSCAN");
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(coords[:, 0], coords[:, 1], c=kmeans_cls)
plt.title("CLS KMEANS");
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(coords[:, 0], coords[:, 1], c=ak_cls)
plt.title("CLS AK");
```
